```python
def clean_dataset(boston_dataframe):
    cleaned_Data = boston_dataframe.dropna()
    return cleaned_Data
```

```python
def  print_histograms(dataframe):
    axes = dataframe.hist(bins=15, figsize=(15, 10))
    plt.tight_layout()
    for ax in axes.flatten():
        ax.set_ylabel("Frequency")
        #ax.set_xlabel("Value")
    plt.show()
```

```python
ld_data = load_dataset()
sumrze_data = print_histograms(ld_data)
```

```python
def compute_correlations_matrix(boston_dataframe):
    corr_matrx = boston_dataframe.corr()
    return corr_matrx

```

```python
from pandas.plotting import scatter_matrix


def print_scatter_matrix(dataframe):
    # Create the scatter matrix plot
    plt.figure(figsize=(15, 15))
    scatter_matrix(dataframe, alpha=0.2, diagonal='hist', figsize=(15, 15), s=30, cmap='viridis')
    
    # Set the title of the plot
    plt.suptitle('Scatter Matrix of Dataset', y=1.02)
    
    # Adjust layout to prevent overlap
    plt.tight_layout()
    
    # Show the plot
    plt.show()
```

```python
boston_dataframe = load_dataset()
scatter_plt =  print_scatter_matrix(ld_data )
print(scatter_plt)
```

```python
import sklearn.linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


def boston_fit_model(boston_dataframe):
  # SELECT two columns from our
  model_dataset = boston_dataframe[["RM","MDEV"]]
  regressor = LinearRegression()
  # Extract column 1
  x = model_dataset.iloc[:, :-1].values
  # Extract column 2
  y = model_dataset.iloc[:, 1].values
  # Train the model
  regressor.fit(x, y)
  return regressor
```

```python
model_dataset = boston_dataframe[["RM","MDEV"]]
fitted_model = boston_fit_model(model_dataset)
print(fitted_model)
```

```python
def boston_predict(estimator, array_to_predict):
    prediction = estimator.predict(array_to_predict)
    return prediction
```
